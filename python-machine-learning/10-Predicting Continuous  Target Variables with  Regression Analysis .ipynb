###Introducing a simple linear regression model

最简单的线性回归模型如下：
$$
y=w_0 + w_1x
$$

多元线性回归
$$
y = w_0x_0 +...+ w_mx_m = \sum_{i=0}^{n}{w_ix_i} = w^Tx
$$
Here,w0 is the y axis intercept with x0 = 1
